# Weather

```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()
random.seed(69)

weather_data = []
cities = ["New York", "London", "Tokyo", "Paris", "Berlin", "Gothenburg", "Singapore"]
conditions = ["sunny", "rainy", "cloudy", "snowy", "foggy"]

for _ in range(50):
    city = random.choice(cities)
    date = (datetime.now() + timedelta(days=random.randint(0,7))).strftime("%Y-%m-%d")
    temp = random.randint(-5, 35)
    condition = random.choice(conditions)
    humidity = random.randint(30, 90)
    wind_speed = random.randint(0, 30)
    
    weather_data.append({
        "city": city,
        "date": date,
        "temperature_min": temp - random.randint(3,5),
        "temperature_max": temp + random.randint(3,5),
        "condition": condition,
        "humidity": humidity,
        "wind_speed": wind_speed
    })

weather_df = pd.DataFrame(weather_data)

weather_df.to_csv("weather_data.csv", index=False)
weather_df.head()
```

# Restaurants

```python
restaurant_data = []
cuisines = ["Italian", "Mexican", "Japanese", "French", "Indian"]
price_ranges = ["$", "$$", "$$$"]

for _ in range(50):
    restaurant_data.append({
        "name": fake.company(),
        "city": random.choice(cities),
        "cuisine": random.choice(cuisines),
        "price_range": random.choice(price_ranges),
        "rating": round(random.uniform(3.0, 5.0), 1),
        "address": fake.address().replace('\n', ', ')
    })

restaurants_df = pd.DataFrame(restaurant_data)

restaurants_df.to_csv("restaurant_data.csv", index=False)
restaurants_df.head()
```

# Public Transport

```python
transport_data = []
routes = ["Red Line", "Blue Line", "Green Line", "Yellow Line"]
transport_types = ["tram", "bus"]

for _ in range(50):
    departure = (datetime.now() + timedelta(minutes=random.randint(5,120))).strftime("%H:%M")
    arrival = (datetime.strptime(departure, "%H:%M") + timedelta(minutes=random.randint(15,45))).strftime("%H:%M")
    
    transport_data.append({
        "route": random.choice(routes),
        "type": random.choice(transport_types),
        "departure_time": departure,
        "arrival_time": arrival,
        "from_stop": fake.street_name(),
        "to_stop": fake.street_name(),
        "city": random.choice(cities)
    })

transport_df = pd.DataFrame(transport_data)

transport_df.to_csv("transport_data.csv", index=False)
transport_df.head()
```
