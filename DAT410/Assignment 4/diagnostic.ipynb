## Classification Function

```python
size_features = ['radius_0', 'radius_2', 'perimeter_0', 'perimeter_2', 'area_0', 'area_2']
shape_features = ['compactness_0', 'compactness_0', 'concavity_0', 'concavity_2', 'concave points_0', 'concave points_2', 'symmetry_2', 'fractal dimension_2']
texture_features = ['texture_0', 'texture_2']
homogeneity_features = [f'{feature}_1' for feature in ['radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness', 'concavity', 'concave points', 'symmetry', 'fractal dimension']]

all_features = size_features + shape_features + texture_features + homogeneity_features

def classify_malignant(patient_features, thresholds):
    """
    0 for mean, 1 for standard error, 2 for worst
    """
    size_abnormal = any(patient_features[f] > thresholds[f] for f in size_features)

    shape_abnormal = any(patient_features[f] > thresholds[f] for f in shape_features)

    texture_abnormal = any(patient_features[f] > thresholds[f] for f in texture_features)

    homogeneity_abnormal = any(patient_features[f] > thresholds[f] for f in homogeneity_features)

    return 1 if (size_abnormal or shape_abnormal or texture_abnormal or homogeneity_abnormal) else 0
```

## Separate into Benign data

```python
benign = wdbc[wdbc['malignant'] == 0]
benign
```

## Rule-based Threshold

```python
thresholds = {}
for feature in all_features:
    mu = benign[feature].mean()
    sigma = benign[feature].std()
    thresholds[feature] = mu + 2 * sigma

thresholds
```

```python
wdbc['predicted'] = wdbc.apply(lambda row: classify_malignant(row, thresholds), axis=1)
wdbc
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

wdbc_true = wdbc['malignant']
wdbc_pred = wdbc['predicted']

accuracy = accuracy_score(wdbc_true, wdbc_pred)
precision = precision_score(wdbc_true, wdbc_pred)
recall = recall_score(wdbc_true, wdbc_pred)
f1 = f1_score(wdbc_true, wdbc_pred)
conf_matrix = confusion_matrix(wdbc_true, wdbc_pred)

# Print results
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print("Confusion Matrix:\n", conf_matrix)
```
